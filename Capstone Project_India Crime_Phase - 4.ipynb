```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

import warnings
warnings.filterwarnings('ignore')
```

```python
#importing district wise crime data
district_crime_data=pd.read_csv(r'https://raw.githubusercontent.com/manishkumbhare/Capstone-Project/main/datasets/01_District_wise_crimes_committed_IPC_2001_2012.csv')
district_crime_data.head()
```

### # Requirement No - 4.1
#### Create 3 clusters as below.
1. Sensitive Area's
2. Moderate Area's
3. Peaceful Area's

#### As we require to do Area wise clustering, So We shall be Choosing some major crimes that defines on the area and which are not domestic crimes like - Dowry, Arson, Breach of trust etc.
#### SO we shall be choosing - Murder, Rape, Kidnapping, Robbery, Burglary, Theft, Roits only.

```python
df=district_crime_data[['STATE/UT','DISTRICT', 'YEAR', 'MURDER', 'RAPE', 'KIDNAPPING & ABDUCTION', 'ROBBERY', 'BURGLARY', 'THEFT', 'RIOTS',]]

df.head()
```

```python
# Creating a new column named 'Total_crime' with the sum of all the select crime type
df['Total_crime']=df[['MURDER','RAPE','KIDNAPPING & ABDUCTION','ROBBERY','BURGLARY','THEFT','RIOTS' ]].sum(axis=1)

df.head()
```

```python
# Selecting 'Total_crime' as features for clustering
features = df[['Total_crime']]  # Add more features as needed

# Standardizing the features
scaler = StandardScaler()
features_standardized = scaler.fit_transform(features)

# Applying k-means clustering
kmeans = KMeans(n_clusters=3, random_state=42)
df['CLUSTER'] = kmeans.fit_predict(features_standardized)

# Defining cluster names
cluster_names = {
    0: 'Sensitive Area',
    1: 'Moderate Area',
    2: 'Peaceful Area',
}

# Maping cluster names to the dataframe
df['CLUSTER_NAME'] = df['CLUSTER'].map(cluster_names)

# Displaying the result
print(df[['STATE/UT', 'DISTRICT', 'YEAR', 'CLUSTER_NAME']])

# Assigning the result to a dataframe 'cluster_result'



```
