### # Requirement No 2.4 :
#### Analysis of area vs overall crime

Area for each state is almost similar in the given data as there is No such demographic changes occured during 2001 to 2012.
So we can take Area for state from any of the year. Lets take it from the latest data we have for year 2011

```python
State_data=pd.read_excel(r'https://github.com/manishkumbhare/Capstone-Project/raw/main/datasets/2011_State_wise_Population_education_area.xlsx')

State_data.head()
```

```python
State_wise_Area=State_data[['State/ Union Territory','Area (In sq. km)', 'Population per sq. km.']]
State_wise_Area.rename(columns={'State/ Union Territory':'STATE/UT'}, inplace=True)
State_wise_Area.head()
```

```python
State_wise_Area = State_wise_Area.sort_values(by='STATE/UT', ascending=True)
State_wise_Area.reset_index(drop=True)
```

```python
# need to have same data fromat so transposing the data
crime_2011 = State_Total_crime.T[2011]
crime_2011=pd.DataFrame(crime_2011)
crime_2011
```

```python
State_wise_Area_crime = pd.merge(State_wise_Area, crime_2011, on='STATE/UT', how='inner')
State_wise_Area_crime.rename(columns={2011: 'Crime'}, inplace=True)
```

```python
State_wise_Area_crime['crime per sq km']=State_wise_Area_crime['Crime'] / State_wise_Area_crime['Area (In sq. km)']
State_wise_Area_crime
```

```python
# Plotting the bar graph
plt.figure(figsize=(12, 6))
plt.bar(State_wise_Area_crime['STATE/UT'], State_wise_Area_crime['crime per sq km'], color='r')
plt.xlabel('STATE/UT')
plt.ylabel('Crime per sq km')
plt.title('State wise Crime per sq km')
plt.xticks(rotation=90)
plt.tight_layout()

# Show the plot
plt.show()
```

### # CONCLUSION :
1. Graph clearly shows UT are having max Area vs Crime rate, where ' Delhi ' , ' Chandigarh ' and ' Pondicherry ' is on Top three

2. If we compare Stat wise then Max Area vs Crime rate, then ' Kerala ', ' West Bengal ' and ' Tamil Nadu ' are on max side
   While, ' Arunachal Pradesh ', ' Nagaland ' and ' Sikkim ' are top three from lower side
